```python
# changing not available and unknown values as nan values in ship mode
# by default it considers few values as NA, but I want to it consider Not-Available, unkowon as NA only
df1=pd.read_csv('orders_data.csv',na_values=['Not Available', 'unknown'])
df1['Ship Mode'].unique()
```

```python
df1.head()
```

# Derive new columns discount , sale price and profit

```python
df1['Sale Price']=df1['List Price']-(df1['List Price']*(df1['Discount Percent']/100))
df1['Sale Price']
```

```python
df1.head(3)
```

```python
df1['Profit']=df1['Sale Price']-df1['cost price']
df1.head(3)
```

```python
df1.rename(columns={'Profit':'Unit_Profit','Sale Price':'Unit_Selling_Proice'},inplace= True)
df1.head()
```

```python
df1.info()
```

```python
df1['Order Date']=pd.to_datetime(df1['Order Date'],format="%d-%m-%Y")
df1['Order Date']
```

```python
df1.info()
```

```python
#dropping few columns
df1.drop(columns=['cost price','List Price','Discount Percent'],inplace=True)
df1.head(2)
```

```python
#calculating total profit in each order
df1['Total_Profit']=df1['Quantity']*df1['Unit_Profit']
df1.head(2)
```

```python
df1=df1.dropna()
df1.shape
```

### Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023

```python
df1.head(2)
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], format='%d-%m-%Y')

# Create df1 as a copy of d

# Extract Month and Year from 'Order Date'
df1['Month'] = df1['Order Date'].dt.month
df1['Year'] = df1['Order Date'].dt.year

# Display first two rows
df1.head(2)
```

```python
a = df1.groupby(['Month', 'year']).sum('Total_sales')['Total_sales'].reset_index()
a
```

```python
pivot_table = a.pivot_table(index='Month', columns='year', values='Total_sales', aggfunc='sum')
pivot_table
```

```python
pivot_table = pivot_table.reset_index()
pivot_table
```

```python
pivot_table['YoY_growth'] = pivot_table[2022] - pivot_table[2023]
pivot_table
```

### Which sub category had highest growth by profit in 2023 compare to 2022

```python
m = df1.groupby(['Sub Category', 'Year']).sum(['Total_Profit'])['Total_Profit'].reset_index()
m
```

```python
n = m.pivot_table(index = 'Sub Category', columns='Year', values = 'Total_Profit').reset_index()
n.head()
```

```python
n['diff'] = n[2023] - n[2022]

n.sort_values(by = ['diff'], ascending=False).head(1)
```
